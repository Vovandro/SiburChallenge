```python
import keras
```

```python
def getModel(input_count = 1, out_count = 4):
    model = keras.Sequential()
    model.add(keras.layers.Dense(20, input_dim=input_count))
    model.add(keras.layers.Dense(30, activation='sigmoid'))
    model.add(keras.layers.Dense(500, activation='relu'))
    model.add(keras.layers.Dense(200, activation='sigmoid'))
    model.add(keras.layers.Dense(100, activation='relu'))
    model.add(keras.layers.Dense(50))
    model.add(keras.layers.Dense(out_count))

    model.compile(optimizer='adam',
                  loss='mean_squared_error',
                  metrics=['mean_squared_error', 'mae'])
    return model
```

```python
models = []

#for col in y_train.columns:
#    print(col)
model = getModel(10, 4)
model.fit(x_train, y_train, epochs=1000, validation_data=(x_valid, y_valid), verbose=0)
#    models.append(model)
```

```python
def applyFirstStepNA(row):
    if row.isna().sum().sum() == 0:
        return row
    
    t = row['timestamp']
    t = [ int(t.year), int(t.month), int(t.day), int(t.hour), int(t.minute), int(t.dayofweek) ]
    pred = model_from_time.predict([t])[0]
    
    i = 0
    for col in df_f.columns:
        if col == 'timestamp':
            continue
            
        if np.isnan(row[col]):
            row[col] = pred[i]
        i += 1
    return row
```
